```python
data=pd.read_csv('my_movies.csv')
data
```

```python
df=pd.get_dummies(data)
df
```

```python
items=apriori(df,min_support=0.2,use_colnames=True)
items
```

```python
rules_lift=association_rules(items,metric='lift',min_threshold=0.5)
rules_lift
```

```python
import matplotlib.pyplot as plt
plt.scatter(data=rules_lift,x='support',y='confidence')
```
